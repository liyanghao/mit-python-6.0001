虽然**计算机只做两件事：执行计算和记住计算的结果，但是它这两件事做的非常好**。典型的放在桌子上或者公文包里的电脑每秒大约执行10亿次计算。很难想象这到底有多快。想象一下：把一个球举到离地面一米高的地方，然后放开手，等这个球落到地上时，你的计算机已经执行了超过10亿个指令。关于内存，一台小型计算机可能大约有数百GB的存储空间。这数百GB的存储有多大？如果1个字节等价于1克，那么100GB大约等于10万吨。作为比较，这相当于15000头非洲大象的重量。

  在人类历史的大部分时间里，计算受限于人类大脑的计算速度和使用人类手指来记录计算结果的能力。这意味着通过计算只能解决一小部分问题。即使有了现代计算机的速度，依然存在超出**现代计算模型**的问题，比如气候变化等。但是**越来越多的问题被证明存在计算解**。我们的希望是：等你读完了这本书，你将会自信地使用**计算思维**去解决在你学习、工作甚至每天的生活中遇到的问题。

什么是**计算思维**？

  所有的知识可分为两类：陈述式和过程式。
  
  陈述式知识由事实语句组成。比如，“x的平方根是满足$y\cdot y=x$的数字y。”这是一条事实语句。不幸的是，它没有告诉我们任何有关如何寻找一个平方根的信息。

  过程式知识是关于“怎么做”的知识。Alexandria的Heron是第一个用文档记录一种计算一个数的平方根的方法的人。他的寻找一个数x的平方根的方法可总结为：
  1. 从一个猜测g开始；
  2. 如果$g\cdot g$跟x足够接近，那么就停止，g就是答案；
  3. 否则，通过对g和x/g求平均来创建新的猜测，即(g+x/g)/2；
  4. 使用新的猜测，我们还叫g，重复之前过程，直到$g\cdot g$跟x足够接近。

  
  比如，考虑寻找25的平方根：
  
  1. 设置g为任意数值，比如3；
  2. 我们判断出$3\cdot 3=9$跟25不是足够接近；
  3. 设置g为(3+25/3)/2=5.67；
  4. 我们判断出$5.67\cdot 5.67=32.15$跟25不是足够接近；
  5. 我们设置g为(5.67+25/5.67)/2=5.04；
  6. 我们判断出$5.04\cdot 5.04=25.4$跟25已足够接近，所以我们停止，宣称5.04是一个充分接近25的平方根的数。

注意到这个方法的描述是一个由**简单步骤**和说明每步何时执行的**控制流**组成的**序列**。像这样的描述就是一个算法。这个算法是猜测-核实算法的示例。它以**容易核实一个猜测是否是好的**为基础。

更正式一点，**一个算法是一个描述了一个计算的有限指令列表，当对输入集执行该计算时，它将通过一组明确定义的状态，最终产生一个输出**。

一个算法有点像烹饪书上的一个食谱：
  
  1. 把奶油蛋羹放在火上。
  2. 搅拌。
  3. 用勺子蘸奶油蛋羹。
  4. 取下勺子，手指穿过勺子背面。
  5. 如果剩下的是干净的路径，把蛋奶冻从火上拿开，让它冷却。
  6. 否则重复。

它包括一些用于决定过程何时完成的测试，以及有关执行指令顺序的说明，有时会根据测试跳转到特定指令。

所以，**在一个机械过程里该如何实现食谱这种想法？**

一种方式是设计一台专门计算平方根的机器。听起来很奇怪，事实上，最早的一批计算机器就是**固定-程序计算机**，意味着它们被用来做非常具体的事情，是求解一个特定的数学问题的工具，比如计算炮弹的轨迹等。首批计算机之一就只会解线性方程组，不能做其他事。Alan Turing的bombe机器就专门设计来破解德军的Enigma密码。一些非常简单的计算机仍使用这种方法。比如，一台手持计算器就是一个固定程序计算机。它能做基本的计算，但不能用做文字处理器或者运行视频游戏。要修改这样的一台机器，需要替换整个电路。

第一台真正意义上的现代计算机是Manchester Mark 1。它跟之前的计算机不同的是：它是一台**存储-程序计算机**。
- 它存储和操作一个指令序列，有执行序列中任意一条指令的组件。
- 通过创建一个**指令集架构**，**将计算细化为指令序列**(即程序)，得到一台极其灵活的机器。
- 通过跟数据一样看待指令，它可以容易地修改程序，在程序的控制下做事。

的确如此，**计算机的核心就变成了一个能执行任意合法指令集的程序(即解释器)**。

因此，可用计算机来计算用基本指令集描述的任何事。

  
程序和其操作的数据都留在内存中。通常，**存在一个指向内存中特定位置的程序计数器，计算从程序计数器指向的那个点处开始执行**。大部分情况，解释器就是简单地走到序列中的下一条指令处。有一些情形，解释器执行一个某个测试，基于那个测试，执行可能跳转到指令序列的其他位置。这就是**控制流**，它对允许编写执行复杂任务的程序来说是必要的。

回到之前的食谱类比，给一组固定的调料集，一个好厨师通过以不同的方式组合它们，可以做出数不清的美味佳肴。类似地，给定一组基础特征，好的程序员可以写出数不清的有用应用。这就使得编程需要令人吃惊的努力。

**为了创建指令序列，我们需要编程语言来描述它们，一种给计算机下达指令的方式**。
    
在1936年，英国数学家Alan Turing描述了一种假想的计算设备，被后人称为**通用图灵机**。这台机器有无限的内存，以磁带的形式呈现：在磁带上你可以写0和1，以及一些简单的用于移动磁带、读写磁带的非常简单的基础指令。Church-Turing测试提出：如果一个函数是可计算的，那么可对一台图灵机变编程来计算它。

在Church-Turing测试中的if是重要的。**并不是所有的问题都有计算解**。比如，图灵展示了：不可能编写一个程序，给定一个任意程序，比如P，该程序输出true当且仅当P将永远运行。这就是著名的**停机问题**。

Church-Turing论文直接导致了**图灵完备性**的概念。一个编程语言是图灵完备的，如果它可被用来模拟一台通用图灵机。**所有的现代编程语言都是图灵完备的**。因此，用一种编程语言编写的程序可被转换成用另一种编程语言编写的程序。当然了，虽然有些事情用一种编程语言做会相对容易些，但是从本质上讲，**所有编程语言的计算能力都是相同的**。

幸运的是，没有程序员是基于图灵的原始指令来构建程序的。相反，现代编程语言提供了一个更大、更方便的原语集。但是，“**编程就是组装操作序列**”这种思想依然没有变。

不管你有有什么样的原语集，不论你用原语编写了什么方法，关于编程最好和最坏的事情都是一样的：计算机将精确地做你告诉它要做的事。这是一件好事，因为这意味着你可以让计算机做各种有趣和有用的事。这是一件坏事，因为当它没有做你想让它做的事时，你通常没人可以责备，除了你自己。

这世上有数百种编程语言。不存在最好的语言。不同的语言适合不同种类的应用。比如，MATLAB是一门良好的操作向量和矩阵的语言。C是一门良好的用于编写控制数据网络的语言。PHP是一门良好的构建网站的语言。Python是一门良好的通用目的语言。

**每门编程语言都有一组原始结构、一种语法、一种静态语义、一种语义**。

用自然语言中的英语类比，英语的原始结构是词汇，语法描述了哪些词汇字符串可组成形式良好的语句，静态语义定义了哪些语句是有意义的，语义确定了那些有意义语句的含义。Python中的原始结构包括字面量(比如数字3.2和字符串"abc")和中缀运算符(比如+和/)。

  一门语言的语法定义了哪些字符串和符号是定义良好的。比如，在英语里，字符串"Cat dog boy"不是语法正确的语句，因为英语的语法不接受<noun><noun><noun>形式的语句。在Python里，原语3.2+3.2是语法形式良好的语句，但语句3.2 3.2不是。

  静态语义定义了哪些语法有效的语句有一个含义。比如，在英语里，字符串"I runs fast"具有<pronoun><verb><adverb>的形式，这是一个语法上可接受的语句。但是，它不是一句正确的英语，因为名词"I"是单数，动词"runs"是复数。这是一个有静态语义错误的示例。在Python里，语句3.2/'abc'是语法正确的语句，具有<literal><operator><literal>的形式，但是产生了一个静态语义错误，因为一个数除以一个字符串没有意义。

  一门语言的语义为每个无静态语义错误的、语法正确的符号字符串关联一个含义。在自然语言里，一条语句的语义可能有歧义。比如，语句"I can not praise this student too highly"可以是褒义，也可以是贬义。编程语言被设计出来以便每个合法的程序有且仅有一种含义。

  虽然语法错误是最常见的错误，但是它们是最没有危险的一类错误。每种严格的编程语言会做完整的语法错误检测，不允许用户执行一个带有一个语法错误的程序。而且，在大多数情况下，语言系统会给出足够清晰的错误位置指示，显然该错误需要被修正。

  跟静态语义错误相关的情形稍微有点复杂。有些编程语言在允许一个程序执行前做了大量的静态语义检查，比如Java。其他编程语言在程序执行前做相对少量的静态语义检查，比如C和Python等。Python在运行一个程序的同时做一定量的语义检查。

  通常不会说一个程序有语义错误。如果一个程序没有语法错误，没有静态语义错误，那么它就有意义，即有语义。当然，那并不是说它有创建者期望它有的语义。当一个程序有不是创建者认为它该有的含义时，可能发生了坏的事情。

  如果一个程序有错误，出现了不在预期内的行为，会发生了什么？
  - 程序可能会崩溃，即停止运行，输出某种显然的它已经做了的指示。在一个定义合适的计算系统里，当一个程序崩溃时，它不会对整个系统有伤害。当然了，一些流行的计算机系统没有这种好的性质。几乎每个使用个人计算机的人都运行过一个程序，该程序可以重启整个电脑。
  - 它也可能继续运行，运行，运行，永不结束。如果你没有一个程序完成任务大约应该花费多久的概念的话，那么这种情形可能很难识别出来。
  - 它也可能运行完成，输出一个可能正确，也可能不正确的答案。

  这些情形中的每个都是坏的，最后一个是最差的。当一个程序似乎在做正确的事情但却没有，坏的事情就会发生：财产可能会有损失，病人可能接受致命剂量的放射治疗，飞机可能会坠毁。

  无论什么时候，编写程序应该以这样的方式：当它不能正确工作时，这是不言而喻的。我们将在整本书中讨论如何编写这样的程序。
